```python
from pydub import AudioSegment
import math
import os

#  Correct file path
input_file = r"D:\DSP Project WAV File\DSP Project WAV File\fa him.wav"

# Folder to save 10-second clips
output_folder = r"D:\DSP Project WAV File\DSP Project WAV File\fa_him"

chunk_length_ms = 1 * 1000   # 10 seconds

# Create output folder
os.makedirs(output_folder, exist_ok=True)

# Load audio
audio = AudioSegment.from_file(input_file)
audio_length_ms = len(audio)

# Number of 10 sec chunks
num_chunks = math.ceil(audio_length_ms / chunk_length_ms)
print(f"Total chunks: {num_chunks}")

# Split and export
for i in range(num_chunks):
    start = i * chunk_length_ms
    end = min(start + chunk_length_ms, audio_length_ms)
    chunk = audio[start:end]

    filename = os.path.join(output_folder, f"clip_{i+1:03}.wav")
    chunk.export(filename, format="wav")
    print("Saved:", filename)

print("Done.")
```
